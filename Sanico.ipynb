```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dp = pd.read_csv('/content/disney_princess_popularity_dataset_300_rows.csv')
print(dp.head())
```

1.

```python
top5 = dp.sort_values(by='PopularityScore', ascending=False).head(5)
print(top5)

colors = ['blue', 'silver', 'pink', 'red', 'green']
barChart = top5.plot(
    kind='bar',
    x='PrincessName',
    y='PopularityScore',
    title='Top 5 Princesses by Popularity Score',
    color=colors
)
barChart.set_xlabel('Princess Name')
barChart.set_ylabel('Popularity Score')

plt.show()
```

2.

```python
outfit = dp['OutfitStyleEra'].value_counts()
print(outfit)

colors = sns.color_palette(palette='pink')
piechart = outfit.plot(kind='pie', autopct='%1.1f%%', title='Outfit Style by Era', colors=colors)
piechart.set_ylabel('')
```

3.

```python
villain = dp['VillainType'].value_counts()
print(villain)

colors = sns.color_palette(palette='Pastel2')
piechart = villain.plot(kind='pie', autopct='%1.1f%%', title='Disney Princess Villain Type', colors=colors)
piechart.set_ylabel('')

```
